```python
from scipy.stats import linregress
```

```python
t=np.array([56.5,113.0,181.0,214.5] )
p=np.array([1,5,20,40] )
plt.plot(t,p,"ro")
```

```python
res= linregress(t,p)

tn=np.arange(50,250,20)
pn=res.intercept + res.slope*tn

plt.figure( figsize=(8,6))

plt.plot(p,t,"*")
plt.plot(pn,tn)
plt.plot (p,t,"bo")
plt.xlabel("$p$")
plt.ylabel("$t$")
plt.grid()
plt.show()
```

```python
#curvas
from scipy.optimize import curve_fit

def acetona (x,a,b,c):
    return a+b*x+c*(x**2)

a,b,c=2.5,1.3,0.5

t=np.array([56.5,113.0,181.0,214.5] )
p=np.array([1,5,20,40] )

(A,B,C),_=curve_fit(acetona,p,t)
print(A,B,C)

#nxdat=np.linspace (0,15,50)
nxdat=np.arange(0,40,.1)
ydatAjuste=acetona(nxdat,A,B,C)

plt.plot(p,t,"o",nxdat,ydatAjuste)
plt.legend(["Datos","Ajuste"],loc="best");
```
