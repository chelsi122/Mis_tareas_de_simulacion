```python
#Ejemplo
f=(x**3)+3*(y**3)+3*(x**2)+3*(y**2)+24
f
```

```python
#gradient
nablaf=[f.diff(var) for var in [x,y]]
nablaf
```

```python
#puntoscriticos
pcs=solve(nablaf)
pcs
```

```python
H=hessian(f,[x,y])
H
```

```python
for pc in pcs:
    eig=H.subs(pc).eigenvals()
    print("Punto critico",pc,"sus eigenvalores son: ",eig)
```
